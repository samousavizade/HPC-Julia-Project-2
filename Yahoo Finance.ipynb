```python
import yfinance as yf
data = pd.DataFrame()
# define the ticker symbol
symbols = ["BTC-USD", 
           "ETH-USD",
           "XRP-USD",
           "SOL-USD",
           "LUNA1-USD",
           "ADA-USD",
           "AVA-USD", 
           "DOT-USD",
           "DOGE-USD",
           "MATIC-USD"
           ]

for symbol in symbols:
    #get data on this ticker
    tickerData = yf.Ticker(symbol)
    #get the historical prices for this ticker
    symbol_close_data = tickerData.history(period="1d", start="2021-1-1", end="2022-01-01")["Close"]
    data[symbol] = symbol_close_data

data
```

```python
data.to_csv("crypto_data.csv")
```
